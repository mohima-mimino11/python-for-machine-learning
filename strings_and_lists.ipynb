# String Methods

```python
string = "I wish you well in Hell! haha! yes hell! haha!"
print(len(string))
processedString = string.lower()
#To find substring
print("well" in processedString)
print("Well" in processedString) #As python is case-sensitive

# To find index of substring
forwardIndex = processedString.find("hell")
print(forwardIndex)
backwardIndex = processedString.rfind("hell")
print(backwardIndex)

# For counting substring ocurrences
count = processedString.count("hell")
print(count)

string2 = string.replace("well", "jail")
print(string)
print(string2)
```

# Splitting , Joining , Formatted String

```python
prompt = "what is Proton ?"

tokens = prompt.split()
print(tokens)
print(type(tokens))
sentence = "-".join(tokens)
print(sentence)
print(type(sentence))
```

# List and Lists Methods

```python
# declaration
numbers = [10, 40, 70, 30, 20]

# access
print(numbers[3])
numbers[3] = 50
print(numbers[3])
newList = numbers[0:3]
print(numbers)
print(newList)

## modifications 

## adding a item 
numbers.append(80)
print(numbers)


## insertion at any position 
numbers.insert(0,45)
print(numbers)


## deleting the last value 
numbers.pop()

print(numbers)

## deleting a specific values 
numbers.remove(100)

print(numbers)
```
