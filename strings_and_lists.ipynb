# String Methods

```python
string = "I wish you well in Hell! haha! yes hell! haha!"
print(len(string))
processedString = string.lower()
#To find substring
print("well" in processedString)
print("Well" in processedString) #As python is case-sensitive

# To find index of substring
forwardIndex = processedString.find("hell")
print(forwardIndex)
backwardIndex = processedString.rfind("hell")
print(backwardIndex)

# For counting substring ocurrences
count = processedString.count("hell")
print(count)

string2 = string.replace("well", "jail")
print(string)
print(string2)
```

# List and Lists Methods

```python
# declaration
numbers = [10, 40, 70, 30, 20]
print(numbers)
# access
print(numbers[3])
numbers[3] = 50
print(numbers[3])
newList = numbers[0:3]
print(numbers)
print(newList)

## modifications 

## adding a item 
numbers.append(80)
print(numbers)


## insertion at any position 
numbers.insert(0,45)
print(numbers)


## deleting the last value 
numbers.pop()

print(numbers)

## deleting a specific values 
numbers.remove(40)

print(numbers)

# ascending sort
numbers.sort()
print(numbers)
# reversing the ascending sort
numbers.reverse()
print(numbers)
```

```python
# descending sort
numbers.sort(reverse=True)

print(numbers)
```

# List as Stack

```python
stack =[]
# pushing in stacks
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)

print(stack)

print(f"top element is : {stack[-1]}")
stack.pop()
print(f"top element is : {stack[-1]}")
```

# List as Queue

```python
queue = []

# pushing in queue
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)
print(queue)

print(f"front element is: {queue[0]}")
queue.pop(0)
print(f"front element is: {queue[0]}")
```
