# AI Mood Detector

```python
s = input()
s.lower()
if ("happy" in s) or  ("joy" in s) or ("smile" in s):
  print("Happy Mood")
elif ("sad" in s) or ("cry" in s) or ("angry" in s):
  print("Sad Mood")
else:
  print("Neutral Mood")
```

# AI Brightness Analyzer

```python
inp = input()
list = inp.split()

sum = 0
numList = [int(n) for n in list]

for i in range(len(numList)):
  sum += numList[i]

averageSum = int(sum / len(numList))

if averageSum < 85:
  print("Dark Image")
elif averageSum >= 85 and averageSum <= 170:
  print("Normal Image")
else:
  print("Bright Image")
```

# AI Bias Detector

```python
inp = input()
predictions = inp.split()
n = len(predictions)
# Defensive: if somehow no tokens, treat as fair (constraints say n>=1 though)
if n == 0:
  print("Fair Model")
  exit()
    
countOfA = predictions.count("A")
countOfB = predictions.count("B")


if (countOfA * 100) > (70 * n) or (countOfB * 100) > (70 * n):
  print("Biased Model")
else:
  print("Fair Model")
```

# AI Keyword Counter

```python
strInput = input()
words = strInput.split()  # Renamed for clarity; split into list of words

# Count occurrences for each keyword
countAI = words.count("ai")
countData = words.count("data")
countModel = words.count("model")
countLearn = words.count("learn")
countTrain = words.count("train")
countNeural = words.count("neural")

# Count how many distinct keywords appear at least once
totalKeywords = 0
if countAI > 0:
    totalKeywords += 1
if countData > 0:
    totalKeywords += 1
if countModel > 0:
    totalKeywords += 1
if countLearn > 0:
    totalKeywords += 1
if countTrain > 0:
    totalKeywords += 1
if countNeural > 0:
    totalKeywords += 1

# Apply the rule
if totalKeywords >= 2:
    print("AI Detected")
else:
    print("Not AI Related")
```
