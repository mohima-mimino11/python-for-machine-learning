# Smart Light Pilot

```python
inp = input();
numbers = inp.split()
brightness = float(numbers[0])
threshold = float(numbers[1])
if (brightness >= threshold):
  print("ON")
else:
  print("OFF")
```

# Sensor Calibrator â€” Min-Max Scaling

```python
inp = input()
numbers = inp.split()
x = float(numbers[0])
min_v = float(numbers[1])
max_v = float(numbers[2])
norm = ((x - min_v)/ (max_v - min_v))
print(f"{norm:.2f}")
```

# Mini Loss Tracker

```python
x = int(input())
target = float(input())
lossSum = 0;
for i in range(x):
  loss = float(input())
  lossSum += loss
average = lossSum / x

if(average <= target):
  print("PASS")
else:
  print("RETRY")
```

# Majority Vote Gate

```python
n = int(input())
yesCount = 0
noCount = 0
for i in range(n):
  vote = input()
  if(vote == 'YES'):
    yesCount+=1
  else:
    noCount+=1
    
if(yesCount >= noCount):
  print("ACCEPT")
else:
  print("REJECT")
```
