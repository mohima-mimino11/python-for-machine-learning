# Smart Light Pilot

```python
inp = input();
numbers = inp.split()
brightness = float(numbers[0])
threshold = float(numbers[1])
if (brightness >= threshold):
  print("ON")
else:
  print("OFF")
```

# Sensor Calibrator â€” Min-Max Scaling

```python
inp = input()
numbers = inp.split()
x = float(numbers[0])
min_v = float(numbers[1])
max_v = float(numbers[2])
norm = ((x - min_v)/ (max_v - min_v))
print(f"{norm:.2f}")
```
