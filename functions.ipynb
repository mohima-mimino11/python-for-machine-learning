### Parameters

```python
def greet(user="guest"):
  print(f"hello {user}")

greet("mimino")
greet()
```

```python
def squareAddition(a, b, c):
  print(f"{a} {b} {c}")
  a = a ** 2
  b = b ** 2
  return a + b + c

# positional arguments
ans = squareAddition(10, 20, 30) + 10
print(ans)
# Keyword arguments
ans = squareAddition(b = 10 , c = 20, a = 30) + 10
print(ans)
```

```python
def sqaureAddition(*args): # for taking variable parameters
  summation = 0
  for i in args:
    i = i ** 2
    summation += i
  return summation

ans = sqaureAddition(10,20, 30, 40, 50)
print(ans)
```

```python
def student(**kwargs): 
  for key,val in kwargs.items():
    print(f"{key} : {val}")
    
student(name = "mimino" , cls =10 , roll=30 , marks = 30.5) 
```

### generator

```python
lst = [x for x in range(500)]
# print(lst)
def dataLoader(chunkSize,lst):
  for i in range(0, len(lst), chunkSize):
    yield lst[i:i+chunkSize]
    
x = dataLoader(5, lst)
print(type(x))
print(next(x))
print(next(x))
```

### LAMBDA Functions


```python
def squareAddition(x,y):
  return x ** 2 + y ** 2

print(squareAddition(2,3))

sqaureAdd = lambda x,y : x ** 2 + y ** 2

print(sqaureAdd(2, 3))
  
```

### map

```python
lst = [1, 2, 3, 4, 5 ,6]

def sqaure(x):
  return x ** 2

lst = list(map(sqaure, lst))

print(lst)
```

```python
#string

string = "hello, world! welcome to the age of AI"

string1 = list(map(str.upper, string))

print(string1)

string1 = "".join(string1)

print(string1)
```

### filter

```python
numbers = [x for x in range(100)]

# even
even = list(filter(lambda x : x % 2 == 0 , numbers))

print(even)

fiftyUpper = list(filter(lambda x : x > 50, numbers))

print(fiftyUpper)
```

```python
data = [0 , 1, '', None, 'Hello', [], [1,2], None, True, False, None]

cleanedData = list(filter(None, data))

print(cleanedData)
```

### Reduce

```python
from functools import reduce

lst = [2, 5, 6, 9, 10]

summation = reduce(lambda x, y : x+y, lst)

print(summation)
```

```python
lst2 = [1, 3, 6, 18, 12]

maxValue = reduce(lambda x , y : x if x > y else y, lst2)
print(maxValue)
```
