#Comparison Operator

```python
x = 10; y = 3

greater_than = x > y
greater_than_equal = 10 >= 10
print(greater_than, greater_than_equal)

less_than = x < y
less_than_equal = 10 <= 10
print(less_than, less_than_equal)

less_than = x < y
less_than_equal = 10 <= 10
print(less_than, less_than_equal)
equal = x == y
print(equal)

not_equal = x != y
print(not_equal)
```
