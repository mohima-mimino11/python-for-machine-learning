# C. Even, Odd, Positive and Negative

```python
n = int(input())

inp = input()

numbers = inp.split()
# print(numbers)
even = 0; odd = 0; positive = 0; negative = 0 
for i in range(n):
  x = int(numbers[i])
  if x > 0:
    positive = positive + 1
  if x < 0:
    negative = negative + 1
  if (x % 2 == 0):
    even = even + 1
  if (x % 2 == 1):
    odd = odd + 1
print("Even: ", even)
print("Odd: ", odd)
print("Positive: ", positive)
print("Negative: ", negative)
    
```
